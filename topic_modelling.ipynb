```python
import os
import nltk
import pandas as pd

import gensim
import gensim.corpora as corpora

from src.constants import *
```

```python
corpus_words = []
corpus_types = []

for filename in os.listdir(ARTICLES_PATH):
    
    entities_path = os.path.join(ARTICLES_PATH, filename, "entities.csv")
    with open(entities_path, "r") as f:
        entities_df = pd.read_csv(f).drop("Unnamed: 0", axis=1)[["Word", "Type"]].astype(str)
    
    corpus_words.append(list(entities_df["Word"].values))   
    corpus_types.append(list(entities_df["Type"].values))

id2word = corpora.Dictionary(corpus_words)    
id2type = corpora.Dictionary(corpus_types)    
```

```python
corpus_words_ids = [id2word.doc2bow(words) for words in corpus_words]
corpus_types_ids = [id2type.doc2bow(types) for types in corpus_types]
```

```python
from pprint import pprint
# number of topics
num_topics = 10
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus_words_ids,
                                       id2word=id2word,
                                       num_topics=num_topics)
# Print the Keyword in the 10 topics
# pprint(lda_model.print_topics())
doc_lda = lda_model[corpus_types_ids]
```

```python
import pickle 
import pyLDAvis
import pyLDAvis.gensim_models

import warnings
warnings.filterwarnings('ignore')

# Visualize the topics
pyLDAvis.enable_notebook()
LDAvis_data_filepath = os.path.join(ROOT_PATH, 'results', 'ldavis_prepared_' + str(num_topics))
print(LDAvis_data_filepath)
if True:
    LDAvis_prepared = pyLDAvis.gensim_models.prepare(lda_model, corpus_words_ids, id2word)
    with open(LDAvis_data_filepath, 'wb') as f:
        pickle.dump(LDAvis_prepared, f)
        
# load the pre-prepared pyLDAvis data from disk
with open(LDAvis_data_filepath, 'rb') as f:
    LDAvis_prepared = pickle.load(f)
pyLDAvis.save_html(LDAvis_prepared, './results/ldavis_prepared_'+ str(num_topics) +'.html')
LDAvis_prepared
```
